# MLP ~ Multiple Layer Perceptron

Introducing a hidden layer to solve the XOR problem. XOR is a famous example introduced in the 1960s to show that a single layer perceptron cannot solve all problems.

The XOR problem is a binary classification problem where the output is 1 if the inputs are different and 0 if they are the same. The XOR problem is not linearly separable, meaning that you cannot draw a straight line to separate the classes in the input space. By introducing a hidden layer, we can create a non-linear decision boundary that can separate the classes in the XOR problem.

- Activation functions

## Architecture

Layer 1: Input → Hidden
- Linear transformation
- ReLU activation

Layer 2: Hidden → Output
- Linear transformation
- Sigmoid activation

**ReLU** (Rectified Linear Unit) introduces non-linearity, allowing the model to learn complex patterns.
function is defined as `f(x) = max(0, x)`, which means it outputs the input directly if it's positive, otherwise it outputs zero.

**Sigmoid** function `1 / (1 + e^{-x})` squashes the output to the range [0, 1], making it suitable for binary classification like XOR, but what's more important is that it allows us to interpret the output as a probability of belonging to the positive class (1).


```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Define the Architecture
class XORModel(nn.Module):
    def __init__(self):
        super(XORModel, self).__init__()
        # Input (2) -> Hidden (2)
        self.hidden = nn.Linear(2, 2) 
        self.relu = nn.ReLU()
        # Hidden (2) -> Output (1)
        self.output = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid() # Squish output to [0, 1]

    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        x = self.sigmoid(x)
        return x
```

```python
import torch.optim as optim

# 1. Define the Architecture
class XORModel(nn.Module):
    def __init__(self):
        super(XORModel, self).__init__()
        # Input (2) -> Hidden (2)
        self.hidden = nn.Linear(2, 2) 
        self.relu = nn.ReLU()
        # Hidden (2) -> Output (1)
        self.output = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid() # Squish output to [0, 1]

    def forward(self, x):
        x = self.hidden(x)
        x = self.relu(x)
        x = self.output(x)
        x = self.sigmoid(x)
        return x
```

```python

# 2. Setup Data (Tensors)
X = torch.tensor([[0,0], [0,1], [1,0], [1,1]], dtype=torch.float32)
Y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)
```

```python

# 3. Initialize Model, Loss, and Optimizer
model = XORModel()
criterion = nn.MSELoss()  # Our Squared Difference friend
optimizer = optim.SGD(model.parameters(), lr=0.1)  # Our Gradient Descent buddy
```

```python

# 4. The Training Loop (The Pipeline)
for epoch in range(1000):
    # Forward Pass
    predictions = model(X)
    loss = criterion(predictions, Y)
    
    # Backward Pass (The "Unwinding")
    optimizer.zero_grad()  # Clear old gradients (don't want to accumulate)
    loss.backward()        # Calculate new gradients via Chain Rule
    optimizer.step()       # Update W and b: W = W - lr * grad

    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python

# 5. Evaluate the Model on XOR Dataset
print("XOR Model Predictions:\n")
print(f"{'Input':<15} {'Expected':<12} {'Predicted':<12}")
print("-" * 40)

model.eval()  # Set model to evaluation mode
with torch.no_grad():  # Disable gradient calculation for inference
    predictions = model(X)
    
for i in range(len(X)):
    input_pair = f"({X[i][0].item():.0f}, {X[i][1].item():.0f})"
    expected = Y[i][0].item()
    predicted = predictions[i][0].item()
    print(f"{input_pair:<15} {expected:<12.4f} {predicted:<12.4f}")
```
